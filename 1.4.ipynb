```python
import math as math
import numpy as np
import sympy as sym
import sympy.physics.mechanics as mech

mech.init_vprinting() #Arrumando o símbolo de derivada para a notação de Newton
```

```python
#Definindo variáveis
l, m, g = sym.symbols('l, m, g', positive = True)
t, lamb = sym.symbols('t, lambda')

#Para o ângulo Theta
the = sym.symbols(r'\theta', cls = sym.Function) #Arrumando o símbolo para teta
the = the(t) # o ângulo theta é dependente do tempo
the_d =  sym.diff(the, t) #Derivada primeira do ângulo theta em relação ao tempo
the_dd =  sym.diff(the_d, t) #Derivada segunda do ângulo theta em relação ao tempo

#Para r
r = sym.symbols('r', cls = sym.Function) #Arrumando o símbolo para r
r = r(t) # r é dependente do tempo
r_d =  sym.diff(r, t) #Derivada primeira de r em relação ao tempo
r_dd =  sym.diff(r_d, t) #Derivada segunda de r em relação ao tempo
```

```python
#Construção da Lagrangiana
#Energia cinética
T = sym.Rational(1,6)*m*r*r*the_d*the_d
T
```

```python
#Energia potencial gravitacional
h = (r/2)*(1-sym.cos(the))
h
```

```python
#Assim, para U
U = m*g*h
U
```

```python
#A lagrangiana é
L = T - U
L
```

```python
#Calculando as equações de vinculo
#termo1  = (d/dt)(dL/r')
#termo2 = (dL/dr)
#Equação de vínculo na forma: termo1 - termo2 = lambda*(dA/dr)
#Ou seja, (d/dt)(dL/r') - (dL/dr) = lambda*(dA/dr)

termo1 = L.diff(r_d).diff(t)
termo1
```

```python
termo2 = L.diff(r)
termo2
```

```python
#Assim, o lado esquerdo da primeira equação de vínculo fica:
euler_lagrange1 = termo1 - termo2
euler_lagrange1
```

```python
#Calculando as equações de vinculo
#termo3  = (d/dt)(dL/θ')
#termo4 = (dL/dθ)
#Equação de vínculo na forma: termo1 - termo2 = lambda*(dA/dθ)
#Ou seja, (d/dt)(dL/θ') - (dL/dθ) = lambda*(dA/dθ)

termo3 = L.diff(the_d).diff(t)
termo3
```

```python
termo4 = L.diff(the)
termo4
```

```python
#Assim, o lado esquerdo da segunda equação de vínculo fica:
euler_lagrange2 = termo3 - termo4
euler_lagrange2
```
