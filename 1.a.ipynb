```python
import math as math
import numpy as np
import sympy as sym
import sympy.physics.mechanics as mech

mech.init_vprinting() #Arrumando o símbolo de derivada para a notação de Newton
```

```python
#Definindo variáveis
l, m, g = sym.symbols('l, m, g', positive = True)
t = sym.symbols('t')

#Para o ângulo Theta
the = sym.symbols(r'\theta', cls = sym.Function) #Arrumando o símbolo para teta
the = the(t) # o ângulo theta é dependente do tempo
the_d =  sym.diff(the, t) #Derivada primeira do ângulo theta em relação ao tempo
the_dd =  sym.diff(the_d, t) #Derivada segunda do ângulo theta em relação ao tempo

#Para as posições de x e y
x, y = sym.symbols('x, y', cls = sym.Function)
x = x(the) # x é dependente do ângulo theta
y = y(the) # y é dependente do ângulo theta
```

Agora precisamos configurar x e y

Ou seja, explicitar a relação de x e y, escrevendo as coordenadas em relação ao ângulo theta

```python
x = l*sym.sin(the)
y = l*sym.cos(the)
```

Assim, fica definido que:

```python
x
```

```python
y
```

Precisamos então definir que a Lagrangiana é igual a L = T - U

Definiremos, portanto, T, U e L

```python
#Definindo a velocidade V
V = sym.symbols('V', cls = sym.Function)
V = V(the)

#Lembrando que: V(x,y) = x(ponto)*i + y(ponto)*j
V = sym.sqrt(sym.Pow(sym.diff(x, t), 2) + sym.Pow(sym.diff(y, t), 2))
```

A velocidade é necessária para o cálculo de T, assim, temos que a velocidade é igual a:

```python
V
```

```python
#A energia cinética pode ser escrita então
T = sym.Rational(1,2)*m*V*V
T.simplify()
```

```python
#A energia potencial pode ser escrita então como: U = m*g*h
#Definindo primeiro o valor da altura h:
h = sym.symbols('h', cls = sym.Function)
h = h(the)

#Temos que h = l(1-cos(theta))
h = l*(1-sym.cos(the))
h
```

#Prosseguindo, para a potencial gravitacional, temos
U = m*g*h
U

```python
#Escrevendo agora a Lagrangiana para o sistema
L = T - U
L.simplify()
```

Equações de Euler-Lagrange

```python
termo1 = L.diff(the.diff(t)).diff(t)
termo1.simplify()
```

```python
termo2 = L.diff(the)
termo2.simplify()
```

```python
sym.solve(sym.Eq(termo1, termo2),the.diff(t,2))
```
